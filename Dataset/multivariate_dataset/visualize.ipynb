```python
# Plot the whole results
import matplotlib.pyplot as plt
#plt.figure()
plt.figure(figsize=(15, 10))
plt.rcParams["figure.figsize"] = (15,10)
#plt.plot(IGBT1['Cycle'], IGBT1['Vceon'], 'darkred', alpha=1, label="IGBT1", linewidth=4)
plt.plot(IGBT2['CycleCount'], IGBT2['V_ce'], 'peru', alpha=1, label="IGBT1", marker='o', linewidth=1, markersize=1)
plt.plot(IGBT3['CycleCount'], IGBT3['V_ce'], 'olivedrab', alpha=1, label="IGBT2", marker='o', linewidth=1, markersize=1)
plt.plot(IGBT4['CycleCount'], IGBT4['V_ce'], 'darkcyan', alpha=1, label="IGBT3", marker='o', linewidth=1, markersize=1)
plt.plot(IGBT5['CycleCount'], IGBT5['V_ce'], 'mediumblue', alpha=1, label="IGBT4", marker='o', linewidth=1, markersize=1)
plt.plot(IGBT1['CycleCount'], IGBT1['V_ce'], 'purple', alpha=1, label="IGBT5", marker='o', linewidth=1, markersize=1)
plt.plot(IGBT6['CycleCount'], IGBT6['V_ce'], 'red', alpha=1, label="IGBT6", marker='o', linewidth=1, markersize=1)

plt.xlabel('Number of Cycles', fontsize =40, fontweight='bold')
plt.ylabel(r"$\mathbf{V_{CE}}$ (V)", fontsize=40, fontweight='bold')
plt.xticks(fontsize=35)
plt.yticks(fontsize=35)
#plt.title("Gaussian Process Regression")
plt.grid(True, linestyle='--')
legend_properties = {
                     'size': 35}
plt.legend(prop=legend_properties)
# Add a thicker border to the plot
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(3)
plt.savefig('/content/drive/MyDrive/Journal-1/Dataset/multivariate_dataset/Vce_vs_Cycle.png',
            dpi=300)
plt.show()
```
